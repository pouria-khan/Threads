```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
# Load your data
posts = pd.read_csv("/Users/pouriakhansari/Library/CloudStorage/OneDrive-IveyBusinessSchool/Prashant/Content Format/Code/EnsembleData/Instagram/ig_post_info.csv")

posts.head()
```

```python
user=pd.read_csv("/Users/pouriakhansari/Library/CloudStorage/OneDrive-IveyBusinessSchool/Prashant/Content Format/Code/EnsembleData/Instagram/ig_user_info.csv")
user
```

```python
df = posts.merge(user,left_on='ig_brand_ID',right_on='user_id',how='left'
)
df
```

```python
df = df[(df["post_type"] != "GraphVideo") & (df["video_count"] == 0) ]
df = df[df['likes'] >= 0]
# difference in seconds
df['age_seconds'] = df['timestamp'] - df['scrape_ts']

# convert to days
df['age_days'] = df['age_seconds'] / 86400

df
```

```python
df_m = df[['likes', 'media_count','comments','post_type','brand_x','followers','age_days', 'business_type']].copy()
df_m['log_followers'] = np.log(df_m['followers'])

df_m = df_m.dropna()
```

```python
model = smf.glm(
    'likes ~ C(post_type) + C(business_type) + log_followers + age_days',
    data=df_m,
    family=sm.families.NegativeBinomial()
).fit()


print(model.summary())
```

```python
model = smf.glm(
    'likes ~ media_count + I(media_count**2) + business_type + log_followers + age_days ',
    data=df_m,
    family=sm.families.NegativeBinomial()
).fit()


print(model.summary())
```

```python
model = smf.glm(
    'comments ~ C(post_type) + business_type + log_followers + age_days ',
    data=df_m,
    family=sm.families.NegativeBinomial()
).fit()


print(model.summary())
```

```python

model = smf.ols(
    'comments ~ C(post_type) + business_type  + log_followers + age_days',
    data=df_m
).fit()

print(model.summary())
```
